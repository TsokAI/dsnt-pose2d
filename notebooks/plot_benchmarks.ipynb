```python
%matplotlib inline

import matplotlib.markers
import matplotlib.pyplot as plt
```

```python
# Benchmark results
points = [
    # (model-name, (time, time-err), accuracy)
    ('resnet18-t1', (4.07, 0.43), 0.803422),
    ('resnet34-t1', (6.35, 0.52), 0.858326),
    ('resnet101-t1', (15.54, 1.07), 0.864188),
    ('hg1', (11.73, 0.66), 0.857489),
    ('hg2', (20.52, 2.04), 0.876703),
]

fig = plt.figure(facecolor='#ffffff')
ax = fig.add_subplot(1, 1, 1)
ax.margins(0.2, 0.2)
ax.set_xlabel('Inference time (ms)')
ax.set_ylabel('Validation accuracy')

colours = plt.get_cmap('Dark2').colors
symbols = matplotlib.markers.MarkerStyle.filled_markers

for (l, (x, xerr), y), c, s in zip(points, colours, symbols):
    ax.errorbar(x, y, xerr=xerr, c=c, marker=s, capsize=5)
    ax.annotate(
        l, xy=(x, y), xytext=(0, 10), color=c,
        textcoords='offset points', ha='center', va='bottom')

matplotlib.rcParams['figure.figsize'] = (10, 6)
display(fig)
plt.close()
```
